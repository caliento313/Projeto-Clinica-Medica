```python
# Sistema simples de cadastro de pacientes - Cl√≠nica Vida+

class Paciente:
    def __init__(self, nome, idade, telefone):
        self.nome = nome
        self.idade = idade
        self.telefone = telefone

    def __str__(self):
        return f"Nome: {self.nome}, Idade: {self.idade}, Telefone: {self.telefone}"

#Fun√ß√£o para nomear os objetos
class Clinica:
    def __init__(self):
        self.pacientes = []

    def cadastrar_paciente(self, nome, idade, telefone):
        paciente = Paciente(nome, idade, telefone)
        self.pacientes.append(paciente)
        print(" Paciente cadastrado com sucesso!")

    def estatisticas(self):
        if not self.pacientes:
            print("Nenhum paciente cadastrado.")
            return

        total = len(self.pacientes)
        idades = [p.idade for p in self.pacientes]
        media = sum(idades) / total
        mais_novo = min(self.pacientes, key=lambda p: p.idade)
        mais_velho = max(self.pacientes, key=lambda p: p.idade)

        print("\nüìä Estat√≠sticas da cl√≠nica:")
        print(f"- Total de pacientes: {total}")
        print(f"- Idade m√©dia: {media:.1f} anos")
        print(f"- Paciente mais novo: {mais_novo.nome} ({mais_novo.idade} anos)")
        print(f"- Paciente mais velho: {mais_velho.nome} ({mais_velho.idade} anos)")

    def buscar_paciente(self, nome):
        encontrados = [p for p in self.pacientes if p.nome.lower() == nome.lower()]
        if encontrados:
            print("\nüîé Paciente encontrado:")
            for p in encontrados:
                print(p)
        else:
            print("Paciente n√£o encontrado.")


def menu():
    clinica = Clinica()

    while True:
        print("\n=== Sistema Cl√≠nica Vida+ ===")
        print("1. Cadastrar paciente")
        print("2. Mostrar estat√≠sticas")
        print("3. Buscar paciente por nome")
        print("0. Sair")

        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            nome = input("Nome: ")
            idade = int(input("Idade: "))
            telefone = input("Telefone: ")
            clinica.cadastrar_paciente(nome, idade, telefone)

        elif opcao == "2":
            clinica.estatisticas()

        elif opcao == "3":
            nome = input("Digite o nome do paciente: ")
            clinica.buscar_paciente(nome)

        elif opcao == "0":
            print("Encerrando o sistema. At√© logo!")
            break

        else:
            print("Op√ß√£o inv√°lida. Tente novamente.")


if __name__ == "__main__":
    menu()
```
